```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/hp/Downloads/DataSets/DataSets/movies.csv")
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()*100/len(df)
```

```python
df['Bond Actor Salary'].fillna(df['Bond Actor Salary'].median(), inplace =True)
```

```python
data = df.groupby('Director')['Box Office'].sum().to_frame()
```

```python
data.reset_index(inplace = True)
```

```python
plt.figure(figsize=(10,6))
plt.bar(data['Director'],data['Box Office'], color = 'goldenrod', edgecolor = 'black', width = 0.68)
plt.xticks(rotation = 90)
dxy = {'family' : 'serif', 'size': 15}
dt = {'family' : 'serif', 'size': 18, 'weight':'bold'}
plt.title("Bar Graph of Box Offices per Director", fontdict= dt)
plt.xlabel("Director Name", fontdict = dxy)
plt.ylabel("Box Office", fontdict = dxy)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.barh(data['Director'],data['Box Office'], color = 'goldenrod', edgecolor = 'black')
dxy = {'family' : 'serif', 'size': 15}
dt = {'family' : 'serif', 'size': 18, 'weight':'bold'}
plt.title("Bar Graph of Box Offices per Director", fontdict= dt)
plt.xlabel("Box Office", fontdict = dxy)
plt.ylabel("Director Name", fontdict = dxy)
plt.show()
```

```python
data_2 = df.groupby('Year')['Bond Actor Salary'].sum().to_frame()
```

```python
data_2.reset_index(inplace = True)
```

```python
plt.figure(figsize=(10,6))
plt.bar(data_2['Year'],data_2['Bond Actor Salary'], color = 'palevioletred', edgecolor = 'black')
dxy = {'family' : 'serif', 'size': 15}
dt = {'family' : 'serif', 'size': 18, 'weight':'bold'}
plt.title("Bond Actor Salary throughout the years", fontdict= dt)
plt.xlabel("Year", fontdict = dxy)
plt.ylabel("Bond Actor Salary", fontdict = dxy)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(data_2['Year'],data_2['Bond Actor Salary'], marker = 'o', color = 'crimson')
dxy = {'family' : 'serif', 'size': 15}
dt = {'family' : 'serif', 'size': 18, 'weight':'bold'}
plt.title("Bond Actor Salary throughout the years", fontdict= dt)
plt.xlabel("Year", fontdict = dxy)
plt.ylabel("Bond Actor Salary", fontdict = dxy)
plt.show()
```
